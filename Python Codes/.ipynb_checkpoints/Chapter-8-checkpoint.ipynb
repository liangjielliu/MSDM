```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor, plot_tree
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
from sklearn.model_selection import train_test_split
from sklearn import tree

hitters = pd.read_csv('pHitters.csv')

# Display the first 10 rows
print("First 10 rows of the Hitters dataset:")
print(hitters.head(10))
```

```python
# Handle missing values by dropping rows with missing Salary, Salary variable has been log-transformed
hitters = hitters.dropna(subset=['Salary'])


# Select predictors and response
X = hitters[['Years', 'Hits']]
y = hitters['Salary']

# Initialize and fit the Decision Tree Regressor
tree_hitters = DecisionTreeRegressor(random_state=42)
tree_hitters.fit(X, y)

# Model summary
print("\nDecision Tree Regressor with 'Years' and 'Hits':")
print(f"Tree Depth: {tree_hitters.get_depth()}")
print(f"Number of Leaves: {tree_hitters.get_n_leaves()}")
print(f"Feature Importances: {dict(zip(X.columns, tree_hitters.feature_importances_))}")

# Plot the regression tree
plt.figure(figsize=(12,8))
plot_tree(tree_hitters, feature_names=X.columns, filled=True, rounded=True)
plt.title("Regression Tree: Salary ~ Years + Hits")
plt.show()
```

```python
# Prune the tree to have only 3 terminal nodes
tree_hitters_pruned = DecisionTreeRegressor(max_leaf_nodes=3, random_state=42)
tree_hitters_pruned.fit(X, y)

# Model summary after pruning
print("\nPruned Decision Tree Regressor with 'Years' and 'Hits':")
print(f"Tree Depth: {tree_hitters_pruned.get_depth()}")
print(f"Number of Leaves: {tree_hitters_pruned.get_n_leaves()}")
print(f"Feature Importances: {dict(zip(X.columns, tree_hitters_pruned.feature_importances_))}")

# Plot the pruned regression tree
plt.figure(figsize=(12,8))
plot_tree(tree_hitters_pruned, feature_names=X.columns, filled=True, rounded=True)
plt.title("Pruned Regression Tree: Salary ~ Years + Hits (3 Terminal Nodes)")
plt.show()
```

```python
# Select more predictors and response
X_more = hitters[['Years', 'Hits', 'RBI', 'Walks', 'PutOuts', 'Runs']]
y = hitters['Salary']

# Initialize and fit the Decision Tree Regressor
tree_hitters_more = DecisionTreeRegressor(random_state=42)
tree_hitters_more.fit(X_more, y)

# Model summary
print("\nDecision Tree Regressor with More Predictors:")
print(f"Tree Depth: {tree_hitters_more.get_depth()}")
print(f"Number of Leaves: {tree_hitters_more.get_n_leaves()}")
print(f"Feature Importances: {dict(zip(X_more.columns, tree_hitters_more.feature_importances_))}")

# Plot the regression tree
plt.figure(figsize=(20,12))
plot_tree(tree_hitters_more, feature_names=X_more.columns, filled=True, rounded=True)
plt.title("Regression Tree: Salary ~ Years + Hits + RBI + Walks + PutOuts + Runs")
plt.show()
```

```python
# Adjust the minimum impurity decrease to control tree growth
# mindev=0.05 in R is roughly analogous to min_impurity_decrease in sklearn
tree_hitters_control = DecisionTreeRegressor(min_impurity_decrease=0.05, random_state=42)
tree_hitters_control.fit(X_more, y)

# Model summary
print("\nDecision Tree Regressor with min_impurity_decrease=0.05:")
print(f"Tree Depth: {tree_hitters_control.get_depth()}")
print(f"Number of Leaves: {tree_hitters_control.get_n_leaves()}")
print(f"Feature Importances: {dict(zip(X_more.columns, tree_hitters_control.feature_importances_))}")

# Plot the regression tree
plt.figure(figsize=(20,12))
plot_tree(tree_hitters_control, feature_names=X_more.columns, filled=True, rounded=True)
plt.title("Regression Tree with min_impurity_decrease=0.05")
plt.show()
```

```python
# Further adjust the minimum impurity decrease
tree_hitters_control_further = DecisionTreeRegressor(min_impurity_decrease=0.005, random_state=42)
tree_hitters_control_further.fit(X_more, y)

# Model summary
print("\nDecision Tree Regressor with min_impurity_decrease=0.005:")
print(f"Tree Depth: {tree_hitters_control_further.get_depth()}")
print(f"Number of Leaves: {tree_hitters_control_further.get_n_leaves()}")
print(f"Feature Importances: {dict(zip(X_more.columns, tree_hitters_control_further.feature_importances_))}")

# Plot the regression tree
plt.figure(figsize=(20,12))
plot_tree(tree_hitters_control_further, feature_names=X_more.columns, filled=True, rounded=True)
plt.title("Regression Tree with min_impurity_decrease=0.005")
plt.show()
```

```python
# Initialize and fit the Decision Tree Regressor with min_samples_split=40
tree_hitters_split = DecisionTreeRegressor(min_samples_split=40, random_state=42)
tree_hitters_split.fit(X_more, y)

# Model summary
print("\nDecision Tree Regressor with min_samples_split=40:")
print(f"Tree Depth: {tree_hitters_split.get_depth()}")
print(f"Number of Leaves: {tree_hitters_split.get_n_leaves()}")
print(f"Feature Importances: {dict(zip(X_more.columns, tree_hitters_split.feature_importances_))}")

# Plot the regression tree
plt.figure(figsize=(20,12))
plot_tree(tree_hitters_split, feature_names=X_more.columns, filled=True, rounded=True)
plt.title("Regression Tree with min_samples_split=40")
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.preprocessing import LabelEncoder

# Suppress any warnings for cleaner output
import warnings
warnings.filterwarnings('ignore')

# Load the Heart dataset
# Replace 'Heart.csv' with the correct path if it's located elsewhere
heart = pd.read_csv('Heart.csv', header=0)

# Display the first 10 rows
print("First 10 rows of the Heart dataset:")
print(heart.head(10))
```

```python
# Display summary statistics
print("\nSummary Statistics:")
print(heart.describe())

# Check for missing values
print("\nMissing Values:")
print(heart.isnull().sum())

# Drop rows with missing values (if any)
heart = heart.dropna()

# Convert categorical variables to 'category' dtype
# Assuming 'AHD', 'Thal', and 'ChestPain' are categorical in R
heart['AHD'] = heart['AHD'].astype('category')
heart['Thal'] = heart['Thal'].astype('category')
heart['ChestPain'] = heart['ChestPain'].astype('category')

# Display data types
print("\nData Types after Conversion:")
print(heart.dtypes)
```

```python
##################### Building and Visualizing the Classification Tree

# Separate features and target
X = heart.drop('AHD', axis=1)
y = heart['AHD']

# Encode categorical target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Encode categorical features (if any)
# Here, 'ChestPain' and 'Thal' might be categorical features
X_encoded = pd.get_dummies(X, columns=['ChestPain', 'Thal'], drop_first=True)

# Initialize the Decision Tree Classifier
# 'gini' is the default criterion, similar to R's 'gini' split in classification trees
tree_heart = DecisionTreeClassifier(random_state=42)

# Fit the model
tree_heart.fit(X_encoded, y_encoded)

# Summary of the tree
print("\nDecision Tree Classifier Summary:")
print(f"Tree Depth: {tree_heart.get_depth()}")
print(f"Number of Leaves: {tree_heart.get_n_leaves()}")
print("Feature Importances:")
for feature, importance in zip(X_encoded.columns, tree_heart.feature_importances_):
    print(f"{feature}: {importance:.4f}")

# Plot the classification tree
plt.figure(figsize=(20,10))
plot_tree(
    tree_heart, 
    feature_names=X_encoded.columns, 
    class_names=label_encoder.classes_.astype(str), 
    filled=True, 
    rounded=True,
    fontsize=12
)
plt.title("Classification Tree: AHD ~ All Features")
plt.show()
```

```python
####################### Pruning the Tree Using Cost Complexity Pruning

# Import necessary library
from sklearn.tree import plot_tree

# Get the effective alphas and the corresponding total leaf impurities
path = tree_heart.cost_complexity_pruning_path(X_encoded, y_encoded)
ccp_alphas, impurities = path.ccp_alphas, path.impurities

# Plot ccp_alphas vs impurities
plt.figure(figsize=(10, 6))
plt.plot(ccp_alphas, impurities, marker='o', drawstyle="steps-post")
plt.xlabel("Effective alpha")
plt.ylabel("Total Impurity of Leaves")
plt.title("Cost Complexity Pruning Path")
plt.show()

# Exclude the maximum alpha (leading to a single leaf)
ccp_alphas = ccp_alphas[:-1]
impurities = impurities[:-1]

# Train a separate tree for each alpha
clfs = []
for ccp_alpha in ccp_alphas:
    clf = DecisionTreeClassifier(random_state=42, ccp_alpha=ccp_alpha)
    clf.fit(X_encoded, y_encoded)
    clfs.append(clf)

# Plot the trees from left (small alpha) to right (large alpha)
n_trees = len(clfs)
plt.figure(figsize=(20, 20))
for idx, clf in enumerate(clfs):
    plt.subplot(n_trees//5 + 1, 5, idx + 1)
    plot_tree(clf, feature_names=X_encoded.columns, class_names=label_encoder.classes_.astype(str),
              filled=True, rounded=True, fontsize=8)
    plt.axis('off')
    plt.title(f"alpha={ccp_alphas[idx]:.4f}")
plt.tight_layout()
plt.show()
```

```python
################## Cross-Validation to Determine Optimal Tree Size
# Initialize lists to store results
alpha_values = ccp_alphas
alpha_scores = []

# Define cross-validation strategy
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

for clf in clfs:
    scores = cross_val_score(clf, X_encoded, y_encoded, cv=cv, scoring='accuracy')
    alpha_scores.append(scores.mean())

# Convert to numpy arrays for easier manipulation
alpha_values = np.array(alpha_values)
alpha_scores = np.array(alpha_scores)

# Find the alpha that gives the highest mean accuracy
optimal_alpha = alpha_values[np.argmax(alpha_scores)]
print(f"\nOptimal ccp_alpha: {optimal_alpha:.5f}")
print(f"Best Cross-Validation Accuracy: {alpha_scores.max():.4f}")
```

```python
###################### Visualization of Cross-Validation Results
# Plot alpha vs accuracy
plt.figure(figsize=(10, 6))
plt.plot(alpha_values, alpha_scores, marker='o')
plt.xlabel("ccp_alpha")
plt.ylabel("Cross-Validated Accuracy")
plt.title("ccp_alpha vs Cross-Validated Accuracy")
plt.axvline(x=optimal_alpha, color='r', linestyle='--', label=f'Optimal alpha = {optimal_alpha:.5f}')
plt.legend()
plt.show()
```

```python
# Choose specific alphas to visualize pruned trees
selected_alphas = [alpha_values[np.argmax(alpha_scores)], alpha_values[len(alpha_values)//2]]

plt.figure(figsize=(20,10))
for idx, alpha in enumerate(selected_alphas, 1):
    clf = DecisionTreeClassifier(random_state=42, ccp_alpha=alpha)
    clf.fit(X_encoded, y_encoded)
    
    plt.subplot(1, len(selected_alphas), idx)
    plot_tree(
        clf, 
        feature_names=X_encoded.columns, 
        class_names=label_encoder.classes_.astype(str), 
        filled=True, 
        rounded=True,
        fontsize=10
    )
    plt.title(f"Pruned Tree\nccp_alpha={alpha:.5f}")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the Heart dataset
# Replace 'Heart.csv' with the correct path if it's located elsewhere
heart = pd.read_csv('Heart.csv', header=0)

# Display the first 10 rows
print("First 10 rows of the Heart dataset:")
print(heart.head(10))

# 2. Data Preprocessing
# Remove rows with missing values
heart = heart.dropna()

# Convert categorical variables to 'category' dtype
categorical_features = ['AHD', 'Thal', 'ChestPain']
for feature in categorical_features:
    heart[feature] = heart[feature].astype('category')

# Encode categorical variables
# Initialize LabelEncoder for the target variable
label_encoder = LabelEncoder()
heart['AHD_encoded'] = label_encoder.fit_transform(heart['AHD'])

# One-hot encode other categorical features
heart_encoded = pd.get_dummies(heart, columns=['Thal', 'ChestPain'], drop_first=True)

# Display the number of features after encoding
print(f"\nNumber of features after encoding: {heart_encoded.shape[1]}")

# 3. Split the data into Training and Test sets
np.random.seed(10)  # Set seed for reproducibility
train_indices = np.random.choice(heart_encoded.index, size=250, replace=False)
test_indices = heart_encoded.index.difference(train_indices)

X_train = heart_encoded.loc[train_indices].drop(['AHD', 'AHD_encoded'], axis=1)
y_train = heart_encoded.loc[train_indices, 'AHD_encoded']
X_test = heart_encoded.loc[test_indices].drop(['AHD', 'AHD_encoded'], axis=1)
y_test = heart_encoded.loc[test_indices, 'AHD_encoded']

print(f"\nTraining set size: {X_train.shape[0]} samples")
print(f"Test set size: {X_test.shape[0]} samples")

# 4. Implement Bagging using RandomForestClassifier
# Set mtry equivalent by setting max_features to the total number of features
mtry = X_train.shape[1]  # Equivalent to mtry=13 in R

# Initialize the Random Forest Classifier
bag_heart = RandomForestClassifier(
    n_estimators=50,          # Number of trees
    max_features=mtry,        # Number of features to consider at each split
    bootstrap=True,           # Enable bagging
    random_state=1            # Seed for reproducibility
)

# Fit the model on the training data
bag_heart.fit(X_train, y_train)

# Display the model summary
print("\nRandom Forest (Bagging) Model:")
print(bag_heart)

# 5. Make Predictions on the Test Set
yhat_bag = bag_heart.predict(X_test)

# 6. Evaluate the Model
# Create a confusion matrix
conf_matrix = confusion_matrix(y_test, yhat_bag)
conf_matrix_df = pd.DataFrame(conf_matrix, index=label_encoder.classes_, columns=label_encoder.classes_)
print("\nConfusion Matrix:")
print(conf_matrix_df)

# Calculate Accuracy
accuracy = accuracy_score(y_test, yhat_bag)
print(f"\nAccuracy on Test Set: {accuracy:.4f}")

# 7. Feature Importances
importances = bag_heart.feature_importances_
feature_names = X_train.columns

# Create a DataFrame for visualization
feature_importances = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
})

# Sort the DataFrame by importance
feature_importances = feature_importances.sort_values(by='Importance', ascending=False)

print("\nFeature Importances:")
print(feature_importances)

# Plot Feature Importances
plt.figure(figsize=(12,8))
sns.barplot(x='Importance', y='Feature', data=feature_importances)
plt.title('Feature Importances from Bagging Model')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split

# Suppress warnings for cleaner output
import warnings
warnings.filterwarnings('ignore')

# Set random seed for reproducibility
np.random.seed(1)

# 1. Load the Heart dataset
# Replace 'Heart.csv' with the correct path if it's located elsewhere
heart = pd.read_csv('Heart.csv', header=0)

# Display the first 10 rows
print("First 10 rows of the Heart dataset:")
print(heart.head(10))

# 2. Data Preprocessing
# Remove rows with missing values
heart = heart.dropna()

# Convert categorical variables to 'category' dtype
categorical_features = ['AHD', 'Thal', 'ChestPain']
for feature in categorical_features:
    heart[feature] = heart[feature].astype('category')

# Encode categorical variables
# Initialize LabelEncoder for the target variable
label_encoder = LabelEncoder()
heart['AHD_encoded'] = label_encoder.fit_transform(heart['AHD'])

# One-hot encode other categorical features
heart_encoded = pd.get_dummies(heart, columns=['Thal', 'ChestPain'], drop_first=True)

# Display the number of features after encoding
print(f"\nNumber of features after encoding: {heart_encoded.shape[1]}")

# 3. Split the data into Training and Test sets
np.random.seed(10)  # Set seed for reproducibility
train_indices = np.random.choice(heart_encoded.index, size=250, replace=False)
test_indices = heart_encoded.index.difference(train_indices)

X_train = heart_encoded.loc[train_indices].drop(['AHD', 'AHD_encoded'], axis=1)
y_train = heart_encoded.loc[train_indices, 'AHD_encoded']
X_test = heart_encoded.loc[test_indices].drop(['AHD', 'AHD_encoded'], axis=1)
y_test = heart_encoded.loc[test_indices, 'AHD_encoded']

print(f"\nTraining set size: {X_train.shape[0]} samples")
print(f"Test set size: {X_test.shape[0]} samples")

# 4. Implement Bagging using RandomForestClassifier
# R's randomForest with mtry=5 corresponds to max_features=5 in scikit-learn
bag_heart = RandomForestClassifier(
    n_estimators=150,          # Number of trees
    max_features=5,            # Number of features to consider at each split
    bootstrap=True,            # Enable bagging
    random_state=1             # Seed for reproducibility
)

# Fit the Random Forest model
bag_heart.fit(X_train, y_train)

# Display the model summary
print("\nRandom Forest (Bagging) Model:")
print(bag_heart)

# 5. Make Predictions on the Test Set
yhat_rf = bag_heart.predict(X_test)

# 6. Evaluate the Model
# Create a confusion matrix
conf_matrix = confusion_matrix(y_test, yhat_rf)
conf_matrix_df = pd.DataFrame(conf_matrix, index=label_encoder.classes_, columns=label_encoder.classes_)
print("\nConfusion Matrix:")
print(conf_matrix_df)

# Calculate Accuracy
accuracy = accuracy_score(y_test, yhat_rf)
print(f"\nAccuracy on Test Set: {accuracy:.4f}")

# 7. Feature Importances
importances = bag_heart.feature_importances_
feature_names = X_train.columns

# Create a DataFrame for visualization
feature_importances = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
})

# Sort the DataFrame by importance
feature_importances = feature_importances.sort_values(by='Importance', ascending=False).reset_index(drop=True)

print("\nFeature Importances:")
print(feature_importances)

# Plot feature importances
plt.figure(figsize=(12,8))
sns.barplot(x='Importance', y='Feature', data=feature_importances, palette='viridis')
plt.title('Feature Importances from Bagging Model')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split

# Suppress warnings for cleaner output
import warnings
warnings.filterwarnings('ignore')

# Set random seed for reproducibility
np.random.seed(10)

# 1. Load the Heart dataset
# Replace 'Heart.csv' with the correct path if it's located elsewhere
try:
    heart = pd.read_csv('Heart.csv', header=0)
except FileNotFoundError:
    print("Error: 'Heart.csv' file not found. Please ensure the file is in the working directory.")
    exit()

# Display the first 10 rows
print("First 10 rows of the Heart dataset:")
print(heart.head(10))

# 2. Data Preprocessing
# Drop rows with missing values
heart_clean = heart.dropna()
print(f"\nNumber of rows after dropping missing values: {heart_clean.shape[0]}")

# Convert categorical variables to 'category' dtype
# Assuming 'AHD', 'Thal', and 'ChestPain' are categorical in R
categorical_features = ['AHD', 'Thal', 'ChestPain']
for feature in categorical_features:
    if feature in heart_clean.columns:
        heart_clean[feature] = heart_clean[feature].astype('category')
    else:
        print(f"Warning: '{feature}' column not found in the dataset.")

# Encode categorical target variable
label_encoder = LabelEncoder()
if 'AHD' in heart_clean.columns:
    y = label_encoder.fit_transform(heart_clean['AHD'])
else:
    print("Error: Target variable 'AHD' not found in the dataset.")
    exit()

# Encode categorical features ('ChestPain' and 'Thal') using one-hot encoding
features_to_encode = ['ChestPain', 'Thal']
for feature in features_to_encode:
    if feature not in heart_clean.columns:
        print(f"Warning: '{feature}' column not found in the dataset.")

X = pd.get_dummies(heart_clean.drop(['AHD'], axis=1), columns=features_to_encode, drop_first=True)

# Display the number of features after encoding
print(f"\nNumber of features after encoding: {X.shape[1]}")

# 3. Split the Data into Training and Test Sets
# Instead of manually sampling, use train_test_split for robustness
# Training size: 250 samples; Test size: remaining
# Ensure that the dataset has at least 250 samples
total_samples = X.shape[0]
if total_samples < 250:
    print(f"Error: The dataset contains only {total_samples} samples, which is less than the required 250 training samples.")
    exit()

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    train_size=250,
    test_size=total_samples - 250,
    random_state=1,
    shuffle=True,
    stratify=y  # Maintains the proportion of classes
)

print(f"\nTraining set size: {X_train.shape[0]} samples")
print(f"Test set size: {X_test.shape[0]} samples")

# 4. Implementing Gradient Boosting
# Initialize the Gradient Boosting Classifier
boost_heart = GradientBoostingClassifier(
    n_estimators=100,        # Number of boosting stages
    max_depth=3,             # Maximum depth of individual regression estimators
    learning_rate=0.1,       # Shrinkage rate
    loss='deviance',         # Loss function to be optimized (logistic regression)
    random_state=1           # Seed for reproducibility
)

# Fit the model on the training data
boost_heart.fit(X_train, y_train)

# Display feature importances
print("\nGradient Boosting Feature Importances:")
for feature, importance in zip(X_train.columns, boost_heart.feature_importances_):
    print(f"{feature}: {importance:.4f}")

# Plot feature importances
plt.figure(figsize=(12,8))
features = pd.Series(boost_heart.feature_importances_, index=X_train.columns)
features = features.sort_values(ascending=False)
sns.barplot(x=features.values, y=features.index, palette='viridis')
plt.title('Feature Importances from Gradient Boosting Model')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()

# 5. Evaluating the Model
# Make predictions on the test set
yhat_boost = boost_heart.predict_proba(X_test)[:,1]  # Probability estimates for the positive class

# Convert probabilities to class predictions based on 0.5 threshold
pred_boost = (yhat_boost > 0.5).astype(int)

# Create a confusion matrix
conf_matrix_boost = confusion_matrix(y_test, pred_boost)
conf_matrix_boost_df = pd.DataFrame(conf_matrix_boost, index=label_encoder.classes_, columns=label_encoder.classes_)
print("\nConfusion Matrix for Gradient Boosting Model:")
print(conf_matrix_boost_df)

# Calculate accuracy
accuracy_boost = accuracy_score(y_test, pred_boost)
print(f"\nAccuracy on Test Set: {accuracy_boost:.4f}")
```
