```python
import numpy as np
import requests
import matplotlib.pyplot as plt

# Step 1: Download and parse G1 dataset
def load_g1_dataset_from_url(url):
    response = requests.get(url)
    lines = response.text.splitlines()
    
    # Read the first line for the number of nodes and edges
    first_line = lines[0].strip().split()
    num_nodes = int(first_line[0])
    num_edges = int(first_line[1])
    
    # Parse edges
    edges = []
    for line in lines[1:]:
        tokens = line.strip().split()
        if len(tokens) == 3:
            i, j, _ = map(int, tokens)
            edges.append((i - 1, j - 1))  # Convert to 0-index
    return num_nodes, edges

# Step 2: Initialize graph and parameters
def initialize_graph(edges, num_nodes):
    J = np.zeros((num_nodes, num_nodes))
    for i, j in edges:
        J[i, j] = -1
        J[j, i] = -1
    return J

def initialize_variables(num_nodes):
    return np.random.uniform(-1, 1, size=num_nodes)

# Step 3: CIM equation implementation
def simulate_cim(J, c, num_steps=500, delta_t=0.01, p=1.0, T=1e-5):
    num_nodes = len(c)
    c_values = []  # Record c_i(t) at each step for plotting
    for _ in range(num_steps):
        noise = np.random.normal(0, np.sqrt(2 * T / delta_t), size=num_nodes)
        c = c + delta_t * (
            (p - 1 - c**2) * c + J @ c + noise
        )
        c_values.append(c.copy())
    return np.array(c_values)

# Step 4: Calculate MAX-CUT and S(t)
def calculate_max_cut(J, c):
    partition = np.sign(c)
    cut_size = 0
    for i in range(len(J)):
        for j in range(i + 1, len(J)):
            if J[i, j] < 0 and partition[i] != partition[j]:
                cut_size += 1
    return cut_size
    
def calculate_S_t(edges, c_values):
    num_steps = len(c_values)
    S_t = np.zeros(num_steps)
    for t, c in enumerate(c_values):
        x = np.sign(c)
        x[x == 0] = 1  # Handle zero values explicitly
        s = 0
        for i, j in edges:
            s += 1 - x[i] * x[j]
        S_t[t] = 0.5 * s
    return S_t

# Main function
def main():
    g1_url = "https://web.stanford.edu/~yyye/yyye/Gset/G1"

    # Load dataset and initialize graph
    num_nodes, edges = load_g1_dataset_from_url(g1_url)
    J = initialize_graph(edges, num_nodes)

    # Initialize variables
    c = initialize_variables(num_nodes)
    
    # (b) Simulate dynamics and calculate MAX-CUT
    c_values = simulate_cim(J, c)
    max_cut_size = calculate_max_cut(J, c_values[-1])
    print(f"Maximum cut size: {max_cut_size}")

    # (c) Plot c_i(t) versus t for 800 variables
    sample_nodes = np.random.choice(num_nodes, size=800, replace=False)
    plt.figure(figsize=(10, 6))
    for node in sample_nodes:
        plt.plot(c_values[:, node], alpha=0.5)
    plt.title("Curves of $c_i(t)$ versus $t$ for 800 variables")
    plt.xlabel("Time step t")
    plt.ylabel("$c_i(t)$")
    plt.show()

    # (d) Calculate and plot S(t) averaged over 10 runs
    num_steps = c_values.shape[0]
    S_t_runs = np.zeros(num_steps)
    for _ in range(10):
        c = initialize_variables(num_nodes)  # Re-initialize
        c_values = simulate_cim(J, c)
        S_t = calculate_S_t(edges, c_values)
        S_t_runs += S_t
    S_t_runs /= 10  # Average over 10 runs

    plt.figure(figsize=(10, 6))
    plt.plot(S_t_runs, label="Average S(t) over 10 runs")
    plt.title("Average cut-size $S(t)$ versus $t$")
    plt.xlabel("Time step t")
    plt.ylabel("$S(t)$")
    plt.legend()
    plt.show()

if __name__ == "__main__":
    main()
```
