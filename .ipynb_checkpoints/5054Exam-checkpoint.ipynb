```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
import torch
import torch.nn as nn
import torch.optim as optim

# 1. Load the data
data = pd.read_csv('cygm.csv')
X = data[['left_eye', 'right_eye', 'nose', 'ear']].values
y = data['class'].values

# Convert y to numerical labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# 2. PCA dimensionality reduction and visualization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_encoded, cmap='viridis')
plt.title('PCA Dimensionality Reduction to 2D')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(scatter, label='Class')
plt.show()

# 3. VAE dimensionality reduction and visualization
class VAE(nn.Module):
    def __init__(self, input_dim, latent_dim):
        super(VAE, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, latent_dim * 2)
        )
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, 32),
            nn.ReLU(),
            nn.Linear(32, 64),
            nn.ReLU(),
            nn.Linear(64, input_dim)
        )

    def reparameterize(self, mu, log_var):
        std = torch.exp(0.5 * log_var)
        eps = torch.randn_like(std)
        return mu + eps * std

    def forward(self, x):
        h = self.encoder(x)
        mu, log_var = torch.chunk(h, 2, dim=1)
        z = self.reparameterize(mu, log_var)
        return self.decoder(z), mu, log_var

# Set up VAE
input_dim = X.shape[1]
latent_dim = 2
vae = VAE(input_dim, latent_dim)
optimizer = optim.Adam(vae.parameters(), lr=1e-3)

# Train VAE
X_tensor = torch.FloatTensor(X_scaled)
epochs = 1000
for epoch in range(epochs):
    optimizer.zero_grad()
    recon_x, mu, log_var = vae(X_tensor)
    recon_loss = nn.MSELoss(reduction='sum')(recon_x, X_tensor)
    kl_div = -0.5 * torch.sum(1 + log_var - mu.pow(2) - log_var.exp())
    loss = recon_loss + kl_div
    loss.backward()
    optimizer.step()

# Get latent representations
vae.eval()
with torch.no_grad():
    _, mu, _ = vae(X_tensor)
X_vae = mu.numpy()

plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_vae[:, 0], X_vae[:, 1], c=y_encoded, cmap='viridis')
plt.title('VAE Dimensionality Reduction to 2D')
plt.xlabel('Latent Dimension 1')
plt.ylabel('Latent Dimension 2')
plt.colorbar(scatter, label='Class')
plt.show()

# 4. K-means clustering on original data and report error rate
kmeans_original = KMeans(n_clusters=3, random_state=42)
kmeans_original.fit(X_scaled)
y_pred_original = kmeans_original.labels_

# Map cluster labels to actual labels
from scipy.stats import mode
labels = np.zeros_like(y_pred_original)
for i in range(3):
    mask = (y_pred_original == i)
    labels[mask] = mode(y_encoded[mask])[0]

error_rate_original = 1 - accuracy_score(y_encoded, labels)
print(f"Error rate on original data: {error_rate_original:.4f}")

# 5. K-means clustering on PCA and VAE reduced data and report error rates
kmeans_pca = KMeans(n_clusters=3, random_state=42)
kmeans_pca.fit(X)
y_pred_pca = kmeans_pca.labels_

labels = np.zeros_like(y_pred_pca)
for i in range(3):
    mask = (y_pred_pca == i)
    labels[mask] = mode(y_encoded[mask])[0]

error_rate_pca = 1 - accuracy_score(y_encoded, labels)
print(f"Error rate on PCA reduced data: {error_rate_pca:.4f}")

kmeans_vae = KMeans(n_clusters=3, random_state=42)
kmeans_vae.fit(X_vae)
y_pred_vae = kmeans_vae.labels_

labels = np.zeros_like(y_pred_vae)
for i in range(3):
    mask = (y_pred_vae == i)
    labels[mask] = mode(y_encoded[mask])[0]

error_rate_vae = 1 - accuracy_score(y_encoded, labels)
print(f"Error rate on VAE reduced data: {error_rate_vae:.4f}")
```

```python
#Q6
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Load the datasets
train_data = pd.read_csv('train_banknote.csv')
test_data = pd.read_csv('test_banknote.csv')

# Separate features and target
X_train = train_data.drop('class', axis=1).values
y_train = train_data['class'].values
X_test = test_data.drop('class', axis=1).values
y_test = test_data['class'].values

# Normalize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Convert to PyTorch tensors
X_train_tensor = torch.FloatTensor(X_train_scaled)
y_train_tensor = torch.FloatTensor(y_train).view(-1, 1)
X_test_tensor = torch.FloatTensor(X_test_scaled)
y_test_tensor = torch.FloatTensor(y_test).view(-1, 1)

# Logistic Regression
class LogisticRegression(nn.Module):
    def __init__(self, input_size):
        super(LogisticRegression, self).__init__()
        self.linear = nn.Linear(input_size, 1)
    
    def forward(self, x):
        return torch.sigmoid(self.linear(x))

# Train Logistic Regression
model_lr = LogisticRegression(X_train.shape[1])
criterion = nn.BCELoss()
optimizer = optim.SGD(model_lr.parameters(), lr=0.01)

epochs = 1000
for epoch in range(epochs):
    optimizer.zero_grad()
    outputs = model_lr(X_train_tensor)
    loss = criterion(outputs, y_train_tensor)
    loss.backward()
    optimizer.step()

# Evaluate Logistic Regression
model_lr.eval()
with torch.no_grad():
    y_pred_lr = model_lr(X_test_tensor).round()
lr_test_error = 1 - accuracy_score(y_test, y_pred_lr.numpy())
print(f"Logistic Regression Test Error: {lr_test_error:.4f}")

# AdaBoost (simple implementation with decision stumps)
class DecisionStump:
    def __init__(self):
        self.feature_index = None
        self.threshold = None
        self.polarity = None

    def fit(self, X, y, sample_weight):
        best_error = float('inf')
        for feature_index in range(X.shape[1]):
            thresholds = np.unique(X[:, feature_index])
            for threshold in thresholds:
                for polarity in [1, -1]:
                    predictions = np.ones(len(y))
                    predictions[X[:, feature_index] * polarity < threshold * polarity] = 0
                    error = np.sum(sample_weight * np.abs(predictions - y))
                    if error < best_error:
                        best_error = error
                        self.feature_index = feature_index
                        self.threshold = threshold
                        self.polarity = polarity

    def predict(self, X):
        predictions = np.ones(len(X))
        predictions[X[:, self.feature_index] * self.polarity < self.threshold * self.polarity] = 0
        return predictions

class AdaBoost:
    def __init__(self, n_estimators=50):
        self.n_estimators = n_estimators
        self.stumps = []
        self.alphas = []

    def fit(self, X, y):
        sample_weight = np.ones(len(X)) / len(X)
        for _ in range(self.n_estimators):
            stump = DecisionStump()
            stump.fit(X, y, sample_weight)
            predictions = stump.predict(X)
            error = np.sum(sample_weight * np.abs(predictions - y)) / np.sum(sample_weight)
            alpha = 0.5 * np.log((1 - error) / error)
            sample_weight *= np.exp(-alpha * y * predictions)
            sample_weight /= np.sum(sample_weight)
            self.stumps.append(stump)
            self.alphas.append(alpha)

    def predict(self, X):
        predictions = np.zeros(len(X))
        for stump, alpha in zip(self.stumps, self.alphas):
            predictions += alpha * stump.predict(X)
        return np.sign(predictions).astype(int)

# Train AdaBoost
adaboost = AdaBoost(n_estimators=50)
adaboost.fit(X_train_scaled, y_train)
y_pred_ab = adaboost.predict(X_test_scaled)
ab_test_error = 1 - accuracy_score(y_test, y_pred_ab)
print(f"AdaBoost Test Error: {ab_test_error:.4f}")

# Support Vector Machine (using PyTorch's implementation)
class SVM(nn.Module):
    def __init__(self, input_size):
        super(SVM, self).__init__()
        self.linear = nn.Linear(input_size, 1)

    def forward(self, x):
        return self.linear(x)

# Train SVM
model_svm = SVM(X_train.shape[1])
criterion = nn.HingeEmbeddingLoss()
optimizer = optim.SGD(model_svm.parameters(), lr=0.01)

epochs = 1000
for epoch in range(epochs):
    optimizer.zero_grad()
    outputs = model_svm(X_train_tensor)
    loss = criterion(outputs * y_train_tensor, torch.ones_like(y_train_tensor))
    loss.backward()
    optimizer.step()

# Evaluate SVM
model_svm.eval()
with torch.no_grad():
    y_pred_svm = model_svm(X_test_tensor).sign()
svm_test_error = 1 - accuracy_score(y_test, y_pred_svm.numpy())
print(f"SVM Test Error: {svm_test_error:.4f}")

# Convolutional Neural Network
class CNN(nn.Module):
    def __init__(self, input_size):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv1d(1, 64, kernel_size=2)
        self.fc1 = nn.Linear(64 * (input_size - 1), 64)
        self.fc2 = nn.Linear(64, 1)

    def forward(self, x):
        x = x.unsqueeze(1)  # Add channel dimension
        x = torch.relu(self.conv1(x))
        x = x.view(x.size(0), -1)
        x = torch.relu(self.fc1(x))
        x = torch.sigmoid(self.fc2(x))
        return x

# Train CNN
model_cnn = CNN(X_train.shape[1])
criterion = nn.BCELoss()
optimizer = optim.Adam(model_cnn.parameters(), lr=0.001)

epochs = 50
for epoch in range(epochs):
    optimizer.zero_grad()
    outputs = model_cnn(X_train_tensor)
    loss = criterion(outputs, y_train_tensor)
    loss.backward()
    optimizer.step()

# Evaluate CNN
model_cnn.eval()
with torch.no_grad():
    y_pred_cnn = model_cnn(X_test_tensor).round()
cnn_test_error = 1 - accuracy_score(y_test, y_pred_cnn.numpy())
print(f"CNN Test Error: {cnn_test_error:.4f}")
```
