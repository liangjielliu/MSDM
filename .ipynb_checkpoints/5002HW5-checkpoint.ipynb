```python
#Q1
#Part A
import requests
import pandas as pd
base_url = "https://data.etagmb.gov.hk"

route_codes = ["12", "11M", "11S", "11B", "11", "104"]
route_ids = {}

for route_code in route_codes:
    region = "NT"  # 假设在新界 (NT)，根据实际需求修改
    response = requests.get(f"{base_url}/route/{region}/{route_code}")
    
    if response.status_code == 200:
        data = response.json().get("data", [])
        if data:
            # 遍历所有匹配的数据，提取 route_id
            route_ids[route_code] = [item.get("route_id") for item in data]

print("All the minibus routes (Route Code and Route ID):")
print(route_ids)

#Part B
north_gate_stop_ids = []
south_gate_stop_ids = []

for route_code, route_id_list in route_ids.items():
    for route_id in route_id_list:
        # 查询 Route Stops (方向为1)
        response = requests.get(f"{base_url}/route-stop/{route_id}/1")
        
        if response.status_code == 200:
            stops = response.json().get("data", {}).get("route_stops", [])
            
            for stop in stops:
                stop_name = stop.get("name_en", "").lower()  # 用英文名字匹配
                stop_id = stop.get("stop_id")
                
                # 根据 stop 名字判断是否是北门或南门
                if "north station" in stop_name:
                    north_gate_stop_ids.append(stop_id)
                elif "south station" in stop_name:
                    south_gate_stop_ids.append(stop_id)
        
        # 查询 Route Stops (方向为2)
        response = requests.get(f"{base_url}/route-stop/{route_id}/2")
        
        if response.status_code == 200:
            stops = response.json().get("data", {}).get("route_stops", [])
            
            for stop in stops:
                stop_name = stop.get("name_en", "").lower()  # 用英文名字匹配
                stop_id = stop.get("stop_id")
                
                # 根据 stop 名字判断是否是北门或南门
                if "north station" in stop_name:
                    north_gate_stop_ids.append(stop_id)
                elif "south station" in stop_name:
                    south_gate_stop_ids.append(stop_id)

# 去重
north_gate_stop_ids = list(set(north_gate_stop_ids))
south_gate_stop_ids = list(set(south_gate_stop_ids))

# 打印结果
print("\nnorth_gate stop_id:", north_gate_stop_ids)
print("south_gate stop_id:", south_gate_stop_ids)

# 构建 route_id -> route_code 的反向映射
id_to_code = {}

for route_code, route_id_list in route_ids.items():
    for route_id in route_id_list:
        id_to_code[route_id] = route_code

# Part C
def get_timetable(direction):
    # 判断方向
    if direction == "N":
        stop_ids = north_gate_stop_ids
    elif direction == "S":
        stop_ids = south_gate_stop_ids
    else:
        print("No such direction")
        return
    
    timetable = []
    
    # 遍历每个 stop_id 查询 ETA
    for stop_id in stop_ids:
        response = requests.get(f"{base_url}/eta/stop/{stop_id}")
        if response.status_code == 200:
            data = response.json().get("data", [])
            
            for route in data:
                if route.get("enabled"):  # 检查 ETA 服务是否可用
                    # 获取 route_id 并通过映射关系还原 route_code
                    route_id = route.get("route_id")
                    route_code = id_to_code.get(route_id, "Unknown")  # 根据映射关系获取 route_code
                    
                    # 获取方向信息 (route_seq) 从 ETA 数据中
                    route_seq = route.get("route_seq", None)  # 从 ETA 数据获取方向
                    
                    # 查询目的地信息
                    dest_en = "Unknown"
                    route_info_response = requests.get(f"{base_url}/route/{route_id}")
                    if route_info_response.status_code == 200:
                        route_info = route_info_response.json().get("data", [])
                        if route_info:
                            directions = route_info[0].get("directions", [])
                            # 根据 route_seq 匹配目的地
                            for direction_info in directions:
                                if direction_info.get("route_seq") == route_seq:
                                    dest_en = direction_info.get("dest_en", "Unknown")
                                    break
                    
                    # 如果目的地包含 "Hong Kong University of Science and Technology"，跳过
                    if "hong kong university of science and technology" in dest_en.lower():
                        continue
                    
                    # 提取到站时间
                    eta_list = route.get("eta", [])
                    if not eta_list:  # 如果 eta 为空，跳过
                        continue
                    
                    for eta in eta_list:
                        eta_time = eta.get("timestamp", "").split("T")[1][:5]  # 提取 HH:MM
                        timetable.append([route_code, route_id, dest_en, eta_time])
    
    # 转换为 Pandas DataFrame
    df = pd.DataFrame(timetable, columns=["Route Code", "Route ID", "Destination", "ETA"])
    
    # 按 ETA 排序
    df = df.sort_values(by="ETA")
    
    # 输出结果
    return df

# 测试函数
df_north = get_timetable("N")
print("\nnorth gate timetable:")
print(df_north)

df_south = get_timetable("S")
print("\nsouth gate timetable:")
print(df_south)
```

```python
# Q2
# Part a
import pandas as pd
import requests
import time
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta

# 配置变量
start_year = 2004
end_year = 2024
base_url = "https://www.hko.gov.hk/cis/dailyExtract/dailyExtract_{year}.xml"

# 通用函数：抓取每日天气数据
def fetch_weather_data(start_year, end_year, base_url):
    all_data = []
    for year in range(start_year, end_year + 1):
        try:
            url = base_url.format(year=year)
            print(f"Fetching data for year {year}")
            response = requests.get(url)
            response.raise_for_status()

            # 提取 JSON 数据
            json_data = response.json()
            for month_data in json_data.get("stn", {}).get("data", []):
                month = month_data["month"]
                for day_data in month_data["dayData"]:
                    if not day_data[0].isdigit():  # 跳过无效行
                        continue
                    date = f"{year}-{month:02d}-{int(day_data[0]):02d}"
                    total_rainfall = 0.02 if day_data[8].strip() == "Trace" else day_data[8].strip() or None
                    all_data.append([
                        date, *day_data[1:8], total_rainfall
                    ])
        except Exception as e:
            print(f"Error fetching data for year {year}: {e}")
        finally:
            time.sleep(1)  # 避免请求过快
    columns = [
        "Date", "Mean Pressure", "Absolute Daily Max Air Temp.",
        "Mean Air Temp.", "Absolute Daily Min Air Temp.",
        "Mean Dew Point", "Mean Relative Humidity", "Mean Amount of Cloud", "Total Rainfall"
    ]
    return pd.DataFrame(all_data, columns=columns)

# Part b：抓取天气警告数据
def fetch_weather_warning(url, start_date="2004-11-01", end_date="2024-10-31"):
    response = requests.get(url)
    all_period = []
    if response.status_code == 200:
        for line in response.text.splitlines():
            line_data = line.split()
            if len(line_data) >= 10:
                start = f"{line_data[0]}-{line_data[1].zfill(2)}-{line_data[2].zfill(2)} {line_data[3].zfill(2)}:{line_data[4].zfill(2)}"
                end = f"{line_data[5]}-{line_data[6].zfill(2)}-{line_data[7].zfill(2)} {line_data[8].zfill(2)}:{line_data[9].zfill(2)}"
                if start_date <= start[:10] <= end_date:
                    all_period.append([start, end])
    return pd.DataFrame(all_period, columns=["Start", "End"])

# Part c：生成警告标记
def generate_daily_flags(df, warning_period, warning_col):
    flags = []
    for date in df["Date"]:
        is_warning = any(
            datetime.strptime(start, "%Y-%m-%d %H:%M") <= date <= datetime.strptime(end, "%Y-%m-%d %H:%M")
            for start, end in warning_period.values
        )
        flags.append(is_warning)
    df[warning_col] = flags

# Part d：绘制对比直方图
def plot_warning_histograms(df, warning_col, title_prefix, colors):
    parameters = [
        "Mean Pressure", "Absolute Daily Max Air Temp.", "Mean Air Temp.",
        "Absolute Daily Min Air Temp.", "Mean Dew Point", "Mean Relative Humidity",
        "Mean Amount of Cloud", "Total Rainfall"
    ]
    fig, axes = plt.subplots(2, 4, figsize=(12, 6))
    axes = axes.flatten()
    for i, param in enumerate(parameters):
        ax = axes[i]
        sns.histplot(df[df[warning_col]][param].dropna().astype(float), label=f'{title_prefix} Warning',
                     color=colors[0], kde=True, bins=20, alpha=0.6, ax=ax)
        sns.histplot(df[~df[warning_col]][param].dropna().astype(float), label=f'No {title_prefix} Warning',
                     color=colors[1], kde=True, bins=20, alpha=0.6, ax=ax)
        ax.set_title(param)
        ax.legend()
    plt.tight_layout()
    plt.show()

# 主流程
if __name__ == "__main__":
    # Part a
    weather_data = fetch_weather_data(start_year, end_year, base_url)
    weather_data["Date"] = pd.to_datetime(weather_data["Date"])
    weather_data = weather_data[(weather_data["Date"] >= "2004-11-01") & (weather_data["Date"] <= "2024-10-31")]
    weather_data.to_csv("hong_kong_weather_data_from_json.csv", index=False)

    # Part b
    very_hot_url = "https://www.hko.gov.hk/dps/wxinfo/climat/warndb/hot.dat?s=63255"
    cold_url = "https://www.hko.gov.hk/dps/wxinfo/climat/warndb/cold.dat?s=91436"
    very_hot_warning = fetch_weather_warning(very_hot_url)
    cold_warning = fetch_weather_warning(cold_url)
    very_hot_warning.to_csv("very_hot_weather_warning.csv", index=False)
    cold_warning.to_csv("cold_weather_warning.csv", index=False)

    # Part c
    generate_daily_flags(weather_data, very_hot_warning, "Very Hot Weather Warning")
    generate_daily_flags(weather_data, cold_warning, "Cold Weather Warning")
    weather_data.to_csv("final_weather_with_warnings.csv", index=False)

    # Part d
    plot_warning_histograms(weather_data, "Very Hot Weather Warning", "Very Hot", ("red", "blue"))
    plot_warning_histograms(weather_data, "Cold Weather Warning", "Cold", ("green", "orange"))
```
