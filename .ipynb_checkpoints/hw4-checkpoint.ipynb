```python
import numpy as np
import matplotlib.pyplot as plt

# Define the target distribution f(x)  exp(-0.5 * x^2)
def target_distribution(x):
    return np.exp(-0.5 * x**2)

# Metropolis algorithm parameters
n_steps = 100000  # Total number of steps
x_values = np.zeros(n_steps)  # Array to store values of x
x_values[0] = 0  # Initial position
step_size = 1.0  # Step size

# Metropolis algorithm
for i in range(1, n_steps):
    # Propose a new position
    x_new = x_values[i-1] + step_size * np.random.normal()
    # Calculate the acceptance ratio
    acceptance_ratio = target_distribution(x_new) / target_distribution(x_values[i-1])
    # Decide whether to accept the new position
    if np.random.rand() < acceptance_ratio:
        x_values[i] = x_new
    else:
        x_values[i] = x_values[i-1]

# Remove the initial part to reach equilibrium distribution (burn-in)
burn_in = 10000
x_values = x_values[burn_in:]

# Calculate the expected values
mean_x = np.mean(x_values)
mean_x2 = np.mean(x_values**2)
mean_x3 = np.mean(x_values**3)
mean_x4 = np.mean(x_values**4)

print("Expected value of x:", mean_x)
print("Expected value of x^2:", mean_x2)
print("Expected value of x^3:", mean_x3)
print("Expected value of x^4:", mean_x4)

# Plot the histogram and compare with the theoretical probability density function
x_range = np.linspace(-4, 4, 100)
pdf_values = target_distribution(x_range)
plt.hist(x_values, bins=50, density=True, alpha=0.6, label="Metropolis Sampling Histogram")
plt.plot(x_range, pdf_values / np.trapz(pdf_values, x_range), label="Theoretical PDF", color="red")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.legend()
plt.title("Metropolis Algorithm Sampling vs Theoretical PDF")
plt.show()
```
